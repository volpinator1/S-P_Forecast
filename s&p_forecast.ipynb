```python
import yfinance as yfin
```

```python
sp500 = yfin.Ticker("^GSPC")
```

```python
sp500 = sp500.history(period="max")
```

```python
sp500 = sp500.tail(10000)
sp500
```

```python
sp500.index
```

```python
sp500.plot.line(y="Close", use_index=True)
sp500.plot.line(y="Close", use_index=True, logy=True)
```

```python
sp500 = sp500.copy()  # Create a copy of the DataFrame to avoid the warning

sp500.loc[:, "Tomorrow"] = sp500["Close"].shift(-1) # shift values back by one for a tomorrow column
sp500
```

```python
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)
```

```python
sp500
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=10, min_samples_split=10, random_state=1) 

#train model with data from up to 100 days ago
train = sp500[:-100]
test = sp500[-100:]

inputs = ["Open", "Close", "Volume", "High", "Low"]
model.fit(train[inputs], train["Target"])
```

```python

from sklearn.metrics import precision_score

predictions = model.predict(test[inputs])
```

```python
import pandas as pd

predictions = pd.Series(predictions, index=test.index)
```

```python
precision_score(test["Target"], predictions)
```

```python
combined = pd.concat([test["Target"], predictions], axis=1)
```

```python
def predict(train, test, inputs, model):
    model.fit(train[inputs], train["Target"])
    predictions = model.predict(test[inputs])
    predictions = pd.Series(predictions, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], predictions], axis=1)
    return combined
```

```python
#Use 0-10y to predict 11th year, then take 11y of data & predict 12th year,...
def backtest(data, model, inputs, start=2500, step=250): 
    all_predictions = []
    
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, inputs, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp500, model, inputs)
```

```python
#we should now add more predictors/inputs into our model to improve it's accuracy
```

```python
horizons = [2, 5, 62]
new_predictors = []

for horizon in horizons:
    rolling_avgs = sp500.rolling(horizon).mean()
    
    ratio_col = f"Close_Ratio_{horizon}" 
    sp500[ratio_col] = sp500["Close"] / rolling_avgs["Close"]
    
    trend_col = f"Trend_{horizon}" #Sum of # of days the stock price went up
    sp500[trend_col] = sp500.shift(1).rolling(horizon).sum()["Target"]
    
    #consider Macd which would be a cross of two different horizon moving averages
    
    new_predictors += [ratio_col, trend_col]
```

```python
sp500 = sp500.dropna()
sp500
```

```python
model2 = RandomForestClassifier(n_estimators=500, min_samples_split=50, random_state=1) 
```

```python
def predict(train, test, inputs, model):
    model2.fit(train[inputs], train["Target"])
    preds = model2.predict_proba(test[inputs])[:,1]
    preds[preds >= 0.7] = 1 #setting a threshold to log predicted green days if model is at least 70% confident
    preds[preds < 0.7] = 0
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
predictions2 = backtest(sp500, model2, new_predictors)
```

```python
predictions2["Predictions"].value_counts() 
#We can see that our predictions threshold of 0.7 reduces our green day predicions amount
```

```python
precision_score(predictions2["Target"], predictions2["Predictions"])
```
